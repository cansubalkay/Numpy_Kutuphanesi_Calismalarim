```python
#RANDOM NUMBERS
#random,randint vb birçok parametre kullanılabilir.
import numpy as np
a=np.random.random((3,2))
print(a)
a1=np.random.randn(1000)
print(a1)
a2=np.random.randint(10,size=(3,3))
print(a2)
a3=np.random.choice(5,size=10)
print(a3)
```

```python
#LINEAR ALGEBRA
#eigenvalues= özdeğerler, eigenvectors= özvektörler
import numpy as np
x=np.array([[1,2],[3,4]])
eigenvalues, eigenvectors = np.linalg.eig(x)
print(eigenvalues)
print(eigenvectors) # column vector
#e_vec * e_val=A * e_vec
```
